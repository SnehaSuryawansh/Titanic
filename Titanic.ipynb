```python
# A new column created named as family size 

df['family_size']=df['Parch'] + df['SibSp']
```

```python
df.sample(5)
```

```python
# Family size if divided in 3 Category ie alone, medium & large

def family_type(number):
    if number==0:
        return "Alone"
    elif number>0 and number<=4:
        return "Medium"
    else:
        return "Large"
```

```python
df['family_type']=df['family_size'].apply(family_type)
```

```python
df.sample(5)
```

```python
# Dropping SiblingSpouse & ParentChild & Family Size. 
```

```python
df.drop(columns=['SibSp','Parch','family_size'],inplace=True)
```

```python
# Detecting Outliers
```

```python
# Handling outliers in age

df= df[df['Age']<(df['Age'].mean()+3*df['Age'].std())]
```

```python
df.shape
```

```python
# outliers for fare column
```

```python
Q1= np.percentile(df['Fare'],25)
Q3= np.percentile(df['Fare'],75)

outlier_low=Q1 - 1.5 * (Q3 - Q1)
outlier_high=Q3 + 1.5 * (Q3 - Q1)

df=df[(df['Fare']>outlier_low) & (df['Fare']<outlier_high)]
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
Pclass = OneHotEncoder(sparse_output= False)
Sex = OneHotEncoder(sparse_output= False)
Embarked = OneHotEncoder(sparse_output= False)
family_type = OneHotEncoder(sparse_output= False)
```

```python
P_ohe= Pclass.fit_transform(df[['Pclass']])
S_ohe= Sex.fit_transform(df[['Sex']])
E_ohe= Embarked.fit_transform(df[['Embarked']])
f_ohe= family_type.fit_transform(df[['family_type']])
```

```python
df_pclass= pd.DataFrame(P_ohe, columns= Pclass.categories_)
df_pclass= df_pclass.iloc[:,1:]
df_pclass
```

```python
df_Sex= pd.DataFrame(S_ohe, columns= Sex.categories_)
df_Sex= df_Sex.iloc[:,1:]
df_Sex
```

```python
df_emb= pd.DataFrame(E_ohe, columns= Embarked.categories_)
df_emb= df_emb.iloc[:,1:]
df_emb
```

```python
df_fam= pd.DataFrame(f_ohe, columns= family_type.categories_)
df_fam= df_fam.iloc[:,1:]
df_fam
```
